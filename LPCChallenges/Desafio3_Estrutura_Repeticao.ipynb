```python
#EX 11
PriceCds = []
qtcds = int(input('Digite a quantidade de cds comprada: '))

for i in range(0, qtcds):
  price = float(input('Digite o preço do Cd: '))
  PriceCds.append(price)

print('O valor total gasto foi de R${}'.format(sum(PriceCds)))
print('O valor médio gasto foi de R${}'.format(sum(PriceCds)/qtcds))
```

```python
#EX 12

num_maior = 0
alt_maior = 0
alt_menor = 2.91
num_menor = 0
qtalunos = int(input('Digite a quantidade total de alunos: '))

for i in range(0, qtalunos):
  aluno = int(input('Digite o número do aluno: '))
  height = float(input('Digite a altura do aluno em centímetros: '))

  if height > alt_maior:
    alt_maior = height
    num_maior = aluno

  if height < alt_menor:
    alt_menor = height
    num_menor = aluno
    

print('A maior altura é de {:.2f}cm do aluno {}'.format(alt_maior, num_maior))
print('A menor altura é de {:.2f}cm do aluno {}'.format(alt_menor, num_menor))
```
