```python
# QUESTÃO 12

print('Interrogatório')

questions = ['Telefonou para a vítima?', 'Esteve no local do crime?', 
             'Mora perto da vítima?', 'Devia para a vítima?', 
             'Já trabalhou com a vítima?']

classif = 0

for i in range(0, 5):
  print(questions[i])
  ans = str(input('Insira a resposta aqui: '))
  if(ans == 'Sim' or ans == 'sim'):
    classif = classif + 1

if(classif == 5):
  print('Assasino!')
elif(classif == 3 or classif == 4):
  print('Cumplice!')
elif(classif == 2):
  print('Suspeito!')
else:
  print('Inocente!')
```
